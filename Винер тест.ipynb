```python
import numpy as np
from scipy.signal import convolve2d as conv2

from skimage import color, data, restoration
import deconv as d
import graphics as g
import metrics as m

astro = color.rgb2gray(data.astronaut())

np.random.seed(seed=0)

psf = np.ones((5, 5)) / 25
astro_blurred = conv2(astro, psf, 'same')
# Add Noise to Image
astro_noisy = astro_blurred.copy()
astro_noisy += (np.random.poisson(lam=25, size=astro.shape) - 10) / 255.
```

```python
wien = d.wiener(astro_noisy, psf)
wien.optimize(astro, m.MAE, n = 50)
g.plot3imgs(astro, wien.recover_image(), astro_noisy, suptitle=' ', figsize=(8, 5))
```

```python
wien = d.wiener(astro_noisy, psf)
wien.optimize(astro, m.MAE, n = 100)
g.plot3imgs(astro, wien.recover_image(), astro_noisy, suptitle=' ', figsize=(8, 5))
```
